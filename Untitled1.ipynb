```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
ar = np.array([4, 5, 1, 2, 6, 10])
```

```python
fig = plt.plot(ar)
```

```python
fig = plt.plot(ar, marker = 2)  # o, *, ^, 1
```

```python
# markersize, marker edge color, marker face color
fig = plt.plot(ar, marker = "^", ms = 10, mec = "red", mfc = "yellow") 
```

```python
# line : line type
# line styes = dotted, dashed
fig = plt.plot(ar, marker = "^", ms = 10, mec = "red", mfc = "yellow", linestyle = "dashed", color = "green", linewidth = 3) 
```

```python
ar = np.array([4, 5, 1, 2, 6, 10])
ar1 = np.array([3, 5, 6, 8, 9])


plt.plot(ar)
plt.plot(ar1, marker = "^", ms = 10, mec = "red", mfc = "yellow", linestyle = "dashed", color = "green", linewidth = 3)
```

```python
# lables:

fig = plt.plot(ar, marker = "^", ms = 10, mec = "red", mfc = "yellow", linestyle = "dashed", color = "green", linewidth = 3)

plt.xlabel("Age")
plt.ylabel("Groth")

plt.title("This is Test title", loc = "right")
```

```python
fig = plt.plot(ar, marker = "^", ms = 10, mec = "red", mfc = "yellow", linestyle = "dashed", color = "green", linewidth = 3)
# plt.grid(axis = "x")
plt.grid(axis = "y")
```

```python
# scatters


var1 = np.array([4, 2, 6, 8, 9])
var2 = np.array([7, 2, 6, 9, 5])

color = np.array(["red", "blue", "orange", "green", "black"])

plt.scatter(var1, var2, c = color)
```

```python
# bar

arr = np.array([4, 6, 7, 9, 10])
arr1 = np.array([6, 8, 2, 5, 7])

plt.bar(arr, arr1, width=0.3)
```

```python
# histogram

# arr = np.array([100, 200, 300, 400])
arr = np.arange(12)
plt.hist(arr)
```

```python
# pie chart

ar = np.array([35, 25, 25, 15])

labelx = ["Apple", "Orange", "Grapes", "Bnana"]

# Explode

xyz = [0.2, 0, 0.1, 0]

plt.pie(ar, labels = labelx, explode = xyz)

```
