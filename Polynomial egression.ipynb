```python
import pandas as pd
import numpy as np
```

```python
x = 6 * np.random.rand(200, 1) - 3
```

```python
y = 0.8 * x ** 2 + 0.9 *x + 2 + np.random.randn(200, 1)


# y = 0.8x^2 +0.9x + 2
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
y
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
```

```python
obj.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pread = obj.predict(x_test)
```

```python
y_pread
```

```python
r2_score(y_test, y_pread)
```

```python
plt.plot(x_train, obj.predict(x_train), color="red")
plt.scatter(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
oj = PolynomialFeatures(degree = 2)
```

```python
x_train
```

```python
x_train_trnform = oj.fit_transform(x_train)
```

```python
x_test_tranform = oj.transform(x_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train_trnform, y_train)
```

```python
y_pre = lr.predict(x_test_tranform)
```

```python
r2_score(y_test, y_pre)
```
