```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1, 12):
    obj = KMeans(n_clusters = i)
    obj.fit_predict(data)
    wcss.append(obj.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 12), wcss, marker="o")
```

```python
km = KMeans(n_clusters = 4)
```
